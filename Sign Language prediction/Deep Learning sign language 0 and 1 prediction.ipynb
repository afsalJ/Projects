## LETS FIND OUT X AND Y FOR THE UPCOMING WORK

```python
X_zero = Xload[204:409]
X_one = Xload[822:1027]

X = np.concatenate((X_zero, X_one), axis =0)
print(X.shape)
```

```python
zeroes = np.zeros(X.shape[0]//2)
ones = np.ones(X.shape[0]//2)
y = np.concatenate((zeroes, ones), axis=0)
print(y.shape)
y=y.reshape(y.shape[0], 1)
print(y.shape)
```

```python
# flattening or reshapping X into a 2 dimensional data

X = X.reshape(X.shape[0], X.shape[1]*X.shape[2])
print(X.shape)
```

```python
# train and test data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15)

print("X_train:", X_train.shape, "\ty_train:", y_train.shape)
print("\nX_test:", X_test.shape, "\ty_test:", y_test.shape)
```

```python
X_train = X_train.T
X_test = X_test.T
y_train = y_train.T
y_test=y_test.T
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

# The above approach wont work for image data or in deep learnig!!

```python
def gradient_descent(w, b, L, X, y):
    m = X.shape[1]
    
    # cost function
    z = np.dot(w.T, X)+b
    prediction = sigmoid(z)
    loss = -y*np.log(prediction) - (1-y) * np.log(1-prediction)
    cost = (np.sum(loss))/m
    
    # find new w (derrivative of w ) and new b (derrivative of b)
    
    dw = np.dot(X, (prediction-y).T) / m
    db = np.sum(prediction - y) / m
    gradients = {'w':w-L*dw, 'b':b-L*db}
    return cost, gradients
```

```python
w = np.full((X_train.shape[0],1),0.01)
b = 0.0
costs = {}
epochs = 150
for i in range(epochs):
    cost, gradient = gradient_descent(w, b, 0.01,X_train,y_train)
    w=gradient["w"]
    b=gradient["b"]
    if(i%10==0):
        costs[i]=cost
        print(i, " : ", cost)
plt.title("COST VARIATION")
plt.plot(costs.keys(), costs.values())
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.show()
```

```python
z = sigmoid(np.dot(w.T, X_test)+b)
prediction = np.zeros((1, X_test.shape[1]))
for i in range(z.shape[1]):
    if z[0, i] <= 0.5:
        prediction[0, i] = 0
    else:
        prediction[0, i] = 1
print("test accuracy : {} %".format(100-np.mean(np.abs(prediction - y_test))*100))
```
