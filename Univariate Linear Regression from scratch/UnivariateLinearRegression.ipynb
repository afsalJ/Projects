```python
def gradient_descent(m, b, points, L):
  m_gradient = 0
  b_gradient = 0
  n=len(points)
  for i in range(len(points)):
    x = points.iloc[i, 0]
    y = points.iloc[i, 1]
    m_gradient+= -(2/n)*(x*(y - (m*x + b)))
    b_gradient+= -(2/n)*(y-(m*x +b))

  return m - m_gradient * L, b - b_gradient * L
```

```python
m = 0
b = 0
L = 0.0001

# Doing without epochs
prev_m = m
prev_b = b
count = 0
flag = True
while(flag):
  m,b=gradient_descent(m,b, data, L)
  if int(prev_m)==int(m) and int(prev_b)==int(b):
    if count == 10:
      flag = False
    else:
      count= count+1
  prev_m = m
  prev_b = b

# Doing with Epochs
# epochs = 300
# for i in range(epochs):
#   if(i%100 == 0):print(i)
#   m,b=gradient_descent(m,b, data, L)


```
